### Outlier removal

```python
def outliers(a):
    t = df[df['Number of bedrooms'] == a]
    arr1 = np.array(t['House price of unit area'])
    # finding the 1st quartile
    q1 = np.quantile(arr1, 0.25)
    
    # finding the 3rd quartile
    q3 = np.quantile(arr1, 0.75)
    med = np.median(arr1)
    
    # finding the iqr region
    iqr = q3-q1
    
    # finding upper and lower whiskers
    upper_bound = q3+(1.5*iqr)
    lower_bound = q1-(1.5*iqr)
    outliers = arr1[(arr1 <= lower_bound) | (arr1 >= upper_bound)]
    print('The following are the outliers in the boxplot','for',a,'bedrooms',':{}'.format(outliers))
outliers(1)
outliers(2)
outliers(3)
i = df[(df['House price of unit area'] == 117.5) | (df['House price of unit area'] == 78.3) | (df['House price of unit area'] == 78) | (df['House price of unit area'] == 73.6)].index
df.drop(i,axis = 0,inplace =True)
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['House price of unit area'],axis = 1)
y = df['House price of unit area']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 52)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

### Normal residual

```python
y_pred = lr.predict(x_test)
residual_error = y_test-y_pred
sns.displot(residual_error,kind = 'kde')
```

```python
stats.probplot(residual_error,dist = 'norm',plot = plt)
plt.show()
```
